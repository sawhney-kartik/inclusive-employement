```python
# Importing libraries
import pandas as pd
import re
import datetime
import numpy as np
from nltk import tokenize
```

```python
# Importing company review files
msft = pd.read_excel('MicrosoftReviews.xlsx')
intel = pd.read_excel('IntelReviews.xlsx')
cisco = pd.read_excel('CiscoReviews.xlsx')
```

```python
# Setting 'Company' column to company name
msft['Company'] = 'Microsoft'
intel['Company'] = 'Intel'
cisco['Company'] = 'Cisco'
```

```python
# Merging all reviews
data = pd.concat([msft, intel, cisco])
# Resetting index
data.reset_index(inplace=True, drop=True)
data
```

```python
# Removing special characters from dates
data['Date'] = data['Date'].apply(lambda x: re.sub(r'[^\w\s]', '', x).strip())
# Changing string dates to datetime format and selecting year
data['Date'] = data['Date'].apply(lambda x: np.NaN if str(x) == 'nan' else datetime.datetime.strptime(str(x), "%b %d %Y").year)
data
```

```python
# Selecting reviews from 2018
data = data.loc[data['Date'] > 2018]
# Resetting index
data.reset_index(drop=True, inplace=True)
data
```

```python
# Checking to see year distribution of reviews
data['Date'].value_counts()
```

```python
# keywords = ['disability', 'disabled person', 'disabilities', 'autism', 'accessibility', 'accessible', 'inclusion', 'inclusivity', 
#             'inclusive', 'disabled employee', 'impaired', 'captioning', 'voice-to-text', 'voice to text',
#             'discriminate', 'discriminating','discrimination', 'equal employment', 'equal opportunity', 'accommodation', 
#             'accommodating', 'diversity', 'diverse', 'ndeam', 'national disability employment awareness month', 'a11y', 'abilities',
#             'paralysed', 'paralyzed', 'paralysis', 'transcribed', 'transcription', 'blind', 'culture'
#            'blindness', 'screen reader', 'representation', 'therapy', 'disorder', 'dei', 'd&i', 'wheelchair', 
#             'hearing', 'neurodiverse', 'deafness']

# Defining list of keywords
keywords = ['disabled people', 'disabled employee', 'disability', 'disabled person', 'disabilities', 'autism', 'autistic',
                   'accessibility', 'accessible', 'visually impaired', 'hearing impaired', 
                   'deaf', 'accommodation', 'accommodations', 'accommodate', 'accommodating', 'paralysis', 
                    'blind', 'blindness','ndeam', 'national disability employment awareness month', 
                   'a11y', 'abilities', 'paralysed', 'paralyzed', 'wheelchair', 
                    'neurodiverse', 'deafness', 'erg', 'employee resource group', 'resource group', 
            'inclusivity', 'inclusion', 'inclusive', 'discriminate', 
                'discriminating', 'diversity', 'diverse', 'representation', 'dei', 'd&i']
```

```python
# Function to identify all keywords in review
def keys(str):
    key = ''
    for i in keywords:
        if re.search(r'\b' + i + r'\b', str.lower()):
            key +=  i + ' ' 
    return key 

# Function to check the relevance of review based on keywords
def isRelevant(str):
    for i in keywords:
        if re.search(r'\b' + i + r'\b', str.lower()):
            return 1 
    return 0
```

```python
# Adding 'Relevance' and 'Keywords' column to the dataframe 
relevance, key = [], []
for i in range(len(data)):
    relevance.append(isRelevant(data.Pros[i]) or isRelevant(data.Cons[i]))
    key.append(keys(data.Pros[i]) or keys(data.Cons[i]))
data['Relevance'] = relevance
data['Keywords'] = key
```

```python
# Checking relevance distribution
data['Relevance'].value_counts()
```

```python
# Selecting relevant reviews
data = data.loc[data['Relevance'] == 1]
data = data.drop(['Relevance'], axis = 1)
data
```

```python
# Adding columns to check if pros/cons are relevant
data['Rel_Pros'] = data['Pros'].apply(lambda x: isRelevant(x))
data['Rel_Cons'] = data['Cons'].apply(lambda x: isRelevant(x))
```

```python
# Taking relevant pros into a 'Pros' dataframe
pros = data.loc[data['Rel_Pros'] == 1]
pros = pros.drop(['Cons', 'Rel_Pros', 'Rel_Cons'], axis = 1)
# Dropping duplicate pros
pros = pros.drop_duplicates(subset=['Pros'])
# Resetting index
pros.reset_index(drop=True, inplace=True)
# Taking relevant pros into a 'Cons' dataframe
cons = data.loc[data['Rel_Cons'] == 1]
cons = cons.drop(['Pros', 'Rel_Pros', 'Rel_Cons'], axis = 1)
# Dropping duplicate cons
cons = cons.drop_duplicates(subset=['Cons'])
# Resetting index
cons.reset_index(drop=True, inplace=True)
```

```python
# Function to remove bullets
def RemBullets(string):
    digits = '1234567890'
    sym = '.)-'
    for x in digits:
        for y in sym:
            z = x+y
            if x+y in string:
                string = string.replace(z, '').strip()
    return string

# Function to find relevant text in reviews
def RelText(string):
    if '\r\n' in string:
        sent = string.split('\r\n')
    else:
        sent = tokenize.sent_tokenize(string)
    res = []
    for i in sent:
        for j in keywords:
            if j in i.lower():
                res.append(i)
    return ' '.join(list(set(res)))
```

```python
# Removing bullets from pros
pros['RelPros'] = pros['Pros'].apply(lambda x: RemBullets(x))
# Finding relevant text in pros
pros['RelPros'] = pros['RelPros'].apply(lambda x: RelText(x))
# Removing first character bullets from pros
pros['RelPros'] = pros['RelPros'].apply(lambda x: x[1:].strip() if (x[0] == '-') | (x[0] == '*') | (x[0] == '+') else x)
# Removing bullets from cons
cons['RelCons'] = cons['Cons'].apply(lambda x: RemBullets(x))
# Finding relevant text in cons
cons['RelCons'] = cons['RelCons'].apply(lambda x: RelText(x))
# Removing first character bullets from cons
cons['RelCons'] = cons['RelCons'].apply(lambda x: x[1:].strip() if (x[0] == '-') | (x[0] == '*') | (x[0] == '+') else x)
```

```python
# Adding <li> tags for HTML
show = []
for i in range(len(pros)):
    keys = pros.Keywords[i].split()
    str = pros.RelPros[i]
    show.append('<li>' + str + '</li>')
#     for i in keys:
#         str = str.replace(i, '<b>' + i + '</b>')
#     if str[:3] == '<b>':
#         str = str[3:].capitalize()
#         show.append('<li><b>' + str + '</li>')
#     else:
#         show.append('<li>' + str.capitalize() + '</li>')
pros['Show'] = show
```

```python
# Adding <li> tags for HTML
show = []
for i in range(len(cons)):
    keys = cons.Keywords[i].split()
    str = cons.RelCons[i]
    show.append('<li>' + str + '</li>')
#     for i in keys:
#         str = str.replace(i, '<b>' + i + '</b>')
#     if str[:3] == '<b>':
#         str = str[3:].capitalize()
#         show.append('<li><b>' + str + '</li>')
#     else:
#         show.append('<li>' + str.capitalize() + '</li>')
cons['Show'] = show
```
